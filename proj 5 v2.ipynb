```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wvalue = pd.read_csv('C:/Users/Sunny/Downloads/worldvalues_data.csv', na_values = {' ','-1','-2','-3','-4','-5','Missing','Not asked in survey','Not applicable','No answer','DonÂ´t know','HT: Dropped out survey'})
```

```python
wvalue.replace(np.NaN, 0, inplace = True)
```

```python
#Using Pearson Correlation
plt.figure(figsize=(50,50))
cor = wvalue.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
corr_matrix = wvalue.corr().abs()
```

```python
# Create correlation matrix
corr_matrix = wvalue.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.70)]

# Drop features 
wvalue.drop(to_drop, axis=1, inplace=True)
```
