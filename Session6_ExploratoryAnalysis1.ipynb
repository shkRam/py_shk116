### Figures and Subplots
Plots in `matplotlib` reside within a `Figure` object. <br>To plot on a figure, we need to create and add one or more **subplots** to it.

Let's create a figure and add four subplots to it.

```python
fig = plt.figure() # create a figure
fig.suptitle('An Empty matplotlib Figure with Four Subplots') # add a centered title

ax1 = fig.add_subplot(2, 2, 1) # add subplot 1
ax2 = fig.add_subplot(2, 2, 2) # add subplot 2
ax3 = fig.add_subplot(2, 2, 3) # add subplot 3
ax4 = fig.add_subplot(2, 2, 4) # add subplot 4
#figure object main plot > then add many subplots, we got 4 area empty so its called sub plot
#we can create plot in these 4 areas
```

If we issue a plotting command now, `matplotlib` draws on the last `figure` and `subplot` used (creating one if necessary), thus hiding the figure and subplot creation. Let's check this out.

```python
fig = plt.figure() # create a figure

ax1 = fig.add_subplot(2, 2, 1) # add subplot 1
ax2 = fig.add_subplot(2, 2, 2) # add subplot 2
ax3 = fig.add_subplot(2, 2, 3) # add subplot 3
ax4 = fig.add_subplot(2, 2, 4) # add subplot 4

# a plotting command
plt.plot(np.random.randn(50).cumsum(), 'g--'); # g-- is a style option that asks matplotlib to plot a green dashed line
#active subplot
```

### Bar Plots
The `plot.bar()` and `plot.barh()` make horizontal and vertical bar plots.

```python
# load the Tips dataset built into the seaborn module
dfTips = sns.load_dataset("tips")
dfTips['tip_pct'] = dfTips.tip/(dfTips['total_bill'] - dfTips.tip) # new column for tipping percentage
dfTips.head()
```

```python
# a usefule recipie for a series's value frequency
countByGender = dfTips.smoker.value_counts()
#value counts how much record you hve in a column

countByGender.plot.bar(rot=0); #rot : rotation is equal to 0
#calling plot method in series as countBygender is series 
```

```python
# bar plots
fig, axes = plt.subplots(2,1); #2 rows 1 column
countByGender.plot.bar(ax= axes[0], color = 'seagreen', alpha = 0.7, rot=0) # vertical bar plot
countByGender.plot.barh(ax= axes[1], color = 'purple', alpha = 0.7) # horizontal bar plot
#barh is horizintal plot

# adjust the spacing between the subplots
plt.subplots_adjust(hspace=0.8)
```

```python
# Average total bill value by day - using aggregation
avgBill = dfTips.groupby('day')['total_bill'].mean()
avgBill.plot.bar(color='skyblue', rot = 0, alpha = 0.7); #avgbill is dataframe so plot works in it
```

```python
# using seaborn (yiu don't need to group by)
ax = sns.barplot(x="day", y="total_bill", ci = False, data=dfTips) #ci : confidnce interval
#set multi parameters at 
ax.set(xlabel='Day', ylabel='Average Total Bill');
```

```python
# Average tip pecentage by day by gender - using aggregation
pivotTable = pd.pivot_table(dfTips, values='tip_pct',
                            index=['day'], columns=['sex'], aggfunc = np.mean)
pivotTable
```

```python
# dodged bar chart
#pivot table is data frame so we can apply plot in this
pivotTable.plot.bar(rot=0, color=['lightseagreen','pink'])
plt.ylabel('Average Tip %')
plt.xlabel('Day')
plt.legend(loc=2); #legend location search on google
```

```python
# stacked bar chart
pivotTable.plot.bar(stacked = True, rot=0, color=['lightseagreen','pink'])
plt.ylabel('Average Tip %')
plt.legend(loc=2);
```

```python
# using seaborn (https://seaborn.pydata.org/generated/seaborn.barplot.html)
ax = sns.barplot(x="day", y="tip_pct", hue = 'sex', hue_order = ['Female', 'Male'], palette=['pink', 'seagreen'], ci = False, data=dfTips) # note the hue parameter
ax.set(xlabel='Day', ylabel='Average Tip %')
plt.legend(loc = 0);
```

### Visualizing the Distribution of a Dataset

#### Histograms  and Density Plots
Histograms is a kind of barplot that gives a discretized display of value frequency. Data points are split into discrete, evenly spaced bins, and the number of data points in each bin is plotted.

```python
#histogram of tipping percentages of the total bill (https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.hist.html#pandas.DataFrame.plot.hist)
dfTips.tip_pct.plot.hist(bins=50);
```

A related plot is a *density plot*, which is formed by computing an estimate of a continuous probablity distribution that might have generated the observed data. 

```python
# Pandas density plot(https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.density.html#pandas.DataFrame.plot.density)
dfTips.tip_pct.plot.density();
#estimate continuous probablity distribution
```

```python
# seaborn distplot (http://seaborn.pydata.org/generated/seaborn.distplot.html)
np.random.seed(0)
x = np.random.randn(100)
ax = sns.distplot(x)
```

```python
# changing the number of bins
sns.distplot(x, bins=20, kde=False);
```

The most familiar way to visualize a bivariate distribution is a scatterplot, where each observation is shown with point at the x and y values.

```python
# scatter plot - pandas (https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.scatter.html#pandas.DataFrame.plot.scatter)
dfTips.plot.scatter(x="total_bill", y="tip", color = 'seagreen');
```

```python
# scatter plot - seaborn (https://seaborn.pydata.org/generated/seaborn.scatterplot.html)
ax = sns.scatterplot(x="total_bill", y="tip", hue='size', size = 'size', data=dfTips)
#size is like 3 people went for the restaurant
```

We can use seaborn's [pairplot](http://seaborn.pydata.org/generated/seaborn.pairplot.html?highlight=plot) to draw scatter plots to show pair-wise relationships and histograms for univariate distributions.

```python
sns.pairplot(dfTips, hue="smoker"); 
#seperate colors for smoking customer
```

```python
# correlation matrix with heatmap
matrix= dfTips[['total_bill','tip', 'size']].corr()
print(matrix)
sns.heatmap(matrix, vmax = .9, square = True, cmap="PiYG"); #heatmap strnght of correlation
```

### Box Plots
A box plot is a method for graphically depicting groups of numerical data through their quartiles. <br>The box extends from the Q1 to Q3 quartile values of the data, with a line at the median (Q2). <br>The whiskers extend from the edges of box to show the range of the data. <br>The position of the whiskers is set by default to 1.5 * IQR (IQR = Q3 - Q1) from the edges of the box. Outlier points are those past the end of the whiskers.

```python
dfTips.boxplot(column=['total_bill','tip']);
```

### Facet Grids and Categorical Data

One way to visualize categorical data with many categorical variables is to use a *facet grid*. Seaborn has a useful `catplot` function that simplifies making many kinds of facet plots.

```python
sns.catplot(x="size", y="tip_pct", hue="sex", data = dfTips[dfTips.tip_pct < 1]);
```

```python
sns.catplot(x='day', y='tip_pct', hue='time', col='smoker', kind='bar',  palette=['seagreen','salmon'], ci=False, data = dfTips[dfTips.tip_pct < 1]);
```

Instead of grouping by `time` by different bar colors within a facet,we can also expand the facet grid by adding one row per `time` value.

```python
sns.catplot(x='day', y='tip_pct', row='time', col='smoker', kind='bar', ci = False, data=dfTips[dfTips.tip_pct < 1]);
#kind chartype box, bar
```

`catplot` supports other plot types such as boxplots.

```python
sns.catplot(x='tip_pct', y='day', row='time', col='smoker', kind='box', data = dfTips[dfTips.tip_pct<0.5]);
```
